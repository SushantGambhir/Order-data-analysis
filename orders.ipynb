```python
# Reload the CSV while treating specific placeholder strings as missing values
# This helps ensure 'Not Available' and 'unknown' are treated as NaN in the DataFrame
df = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
```

```python
# Convert all column names to lowercase so they are easier to work with consistently
df.columns = df.columns.str.lower()
```

```python
# Display the column names after converting them to lowercase to confirm the rename succeeded
df.columns
```

```python
# Replace spaces in column names with underscores for safer programmatic access
df.columns = df.columns.str.replace(' ','_')
```

```python
# Show the updated column names after replacing spaces with underscores in the headers
df.columns
```

```python
# Now we shall derive new columns: discount , sale price and profit
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
df
```

```python
df['sale_price']= df['list_price']-df['discount']
df
```

```python
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
df.dtypes
```

```python
# Now we want to convert order_date into date-time 
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Now we shall drop cost price, list price and discount percent columns
# since we have derived dicount, profit, sales price columns
df = df.drop(columns=['list_price','cost_price','discount_percent'])

# Alternatively we could have used this statement instead:
# df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
# Create a SQLAlchemy engine to connect to a SQL Server database named TRAINING
# on the local SQLEXPRESS instance using pyodbc.
# - driver=ODBC Driver 18 for SQL Server → specifies the ODBC driver
# - trusted_connection=yes → uses Windows Authentication (no username/password)
# - TrustServerCertificate=yes → skips SSL certificate validation (useful for local/dev setups)
import sqlalchemy as sal
engine = sal.create_engine(
    "mssql+pyodbc://@localhost\\SQLEXPRESS/TRAINING"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&trusted_connection=yes"
    "&TrustServerCertificate=yes"
)
```

```python
engine
```

```python
# Create an active database connection from the SQLAlchemy engine
# The engine manages the connection pool, and this call retrieves a connection
# that can be used to run SQL queries or transactions
conn = engine.connect()
```
